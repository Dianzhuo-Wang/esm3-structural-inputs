```python
import pickle
import numpy as np
import pandas as pd
```

```python
with open("embedding_bloom_ACE2_emb_esm1v_650m.pkl", 'rb') as f:
    data = pickle.load(f)
```

```python
data[0].keys()
```

```python
sequences = [data[i]["sequence"] for i in range(len(data))]
```

```python
len(sequences), len(sequences[0])
```

```python
unique_aa = np.sort(np.unique([seq[0] for seq in sequences[:25]]))
aa_to_index = {aa: index for index, aa in enumerate(unique_aa)}
```

```python
index_variant = 11 # in bloom dataset
wt_sequence = "NITNLCPFGEVFNATRFASVYAWNRKRISNCVADYSVLYNSASFSTFKCYGVSPTKLNDLCFTNVYADSFVIRGDEVRQIAPGQTGKIADYNYKLPDDFTGCVIAWNSNNLDSKVGGNYNYLYRLFRKSNLKPFERDISTEIYQAGSTPCNGVEGFNCYFPLQSYGFQPTNGVGYQPYRVVVLSFELLHAPATVCGPKKST"
wt_sequence_encoding = np.array([aa_to_index[a] for a in wt_sequence])
```

```python
one_hot_encoding = np.stack([[aa_to_index[a] for a in seq] for seq in sequences])
```

```python
nb_of_mutations_per_variant = np.sum(((one_hot_encoding - wt_sequence_encoding.reshape(1, -1)) != 0).astype(int), axis=1)
```

```python
with open("embedding_bloom_ACE2_emb_esm2_650m.pkl", 'rb') as f:
    data_esm2 = pickle.load(f)

mean_representation_esm2 = torch.tensor(np.stack([data_esm2[i]["mean_representation"] for i in range(len(data_esm2))]))
torch.save(mean_representation_esm2, "embeddings_sequence_esm2_650m.pt")
```
